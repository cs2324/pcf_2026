```python
x = sp.Symbol("x")
y = sp.Symbol("y")
i = complex(0, 1)
M1 = sp.Matrix([[x * i, 1, 0], [-1, y * i, 0], [0, 0, 0]])
M2 = sp.Matrix([[0, 0, 0], [0, x * i, 1], [0, -1, y * i]])

def func(S):
    M = sp.eye(3)
    for s in S:
        if s == "1":
            M = M * M1
        elif s == "2":
            M = M * M2
    return sp.expand(M)[0, 2]

func("11112222")
```

```python
def matrix_prod(S):
    M = sp.eye(3)
    for s in S:
        if s == "1":
            M = M * M1
        elif s == "2":
            M = M * M2
    return sp.expand(M)
```

```python
matrix_prod("122112")
```
